```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
from matplotlib.pyplot import figure


# set the chained_assignment option
pd.options.mode.chained_assignment = None

# creating data frame and reading data 
df = pd.read_csv(r'/Users/caaturday/Documents/Python/Movies Project/movies.csv')
```

```python
#----------------- data cleaning 
'''checking to see if there is any missing data'''
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# replace Nans with zeros on the following columns
df['rating'] = df['rating'].replace(np.nan,0)
df['budget'] = df['budget'].replace(np.nan,0)
df['gross'] = df['gross'].replace(np.nan,0)

# will print false if nans were successfully removed 
check_for_nan = df['rating'].isnull().values.any()
print (check_for_nan)
check_for_nan = df['budget'].isnull().values.any()
print (check_for_nan)
check_for_nan = df['gross'].isnull().values.any()
print (check_for_nan)
```

```python
# checking data types
print(df.dtypes)

# changing 'released' from an obj to a str to access data in obj
df['released'] = df['released'].astype(str).str[-1]
```
